## Exercise 2: Running SQL Commands

Azure Databricks notebooks natively support SQL alongside Python. You can switch the language of any cell using **magic commands** — special directives that start with `%`. For example, `%sql` tells Databricks to interpret the cell as SQL.

**Scenario:** The CityMoves Transit operations team works primarily with SQL. You want to demonstrate that SQL works seamlessly in the same notebook where you write Python.

**Your task:** In the cell below, use the `%sql` magic command to write a SQL statement that:

1. Returns the current date and time.
2. Returns a label such as `'CityMoves Transit'` as a column called `system_name`.

The result should be a single-row table with at least two columns.



# Code Explanation

```sql
%sql

SELECT current_timestamp() AS current_time,
       'CityMoves Transit' AS system_name
```

### Line 1
```sql
%sql
```
- `%sql` is a **magic command** used in notebooks such as Databricks.
- It tells the notebook to execute the following statements as **SQL code**.

---

### Line 2
```sql
SELECT current_timestamp() AS current_time,
       'CityMoves Transit' AS system_name
```

#### `SELECT`
- Used to retrieve data or expressions.

#### `current_timestamp()`
- A built-in SQL function.
- Returns the current date and time of the system.

#### `AS current_time`
- `AS` assigns an alias (temporary column name).
- The result of `current_timestamp()` will be shown under the column name **current_time**.

#### `'CityMoves Transit'`
- A string literal enclosed in single quotes.

#### `AS system_name`
- Assigns the alias **system_name** to the string value.

---

### Output

| current_time | system_name |
|-------------|-------------|
| 2026-06-25 10:30:15 | CityMoves Transit |

### Key Concepts
- `%sql` → Run SQL code in a notebook.
- `SELECT` → Retrieve values.
- `current_timestamp()` → Get current date and time.
- `AS` → Rename a column in the result set.
- `'text'` → String literal in SQL.

````python
%sql


SELECT current_timestamp() AS current_time, 'CityMoves Transit' AS system_name
````
